## Análisis de valores nulos

```python
def get_null_df_analisis(df):
    nullDF = pd.DataFrame()
    for c_in in df.columns:
        sum_nulls = np.sum(pd.isnull(df[c_in]))
        r = pd.DataFrame({
            "Columna": [c_in],
            "SumaNulos": [sum_nulls],
            "porcentaje": [sum_nulls / df.shape[0]]
        })
        nullDF = pd.concat([nullDF, r])
    return nullDF
```

```python
def print_bars( df , col, var, tit="", ylab="" , xlab=""):
    x  = df.groupby( [col] , as_index = False ).agg( { var : ["count"] } ) 
    x.columns = x.columns.droplevel(1)
    x.columns = [col,'count']
    x = x.sort_values(["count"], ascending=False)
    plt.barh( x[col] , x['count'], align='center')
    plt.title(tit,fontsize=20)
    plt.xlabel(xlab,fontsize=12)
    plt.ylabel(ylab,fontsize=12)
```

```python
def get_freq_cleaning_characters_genero(df, col, gen,sep=";"):
    summary = pd.DataFrame()
    for index, row in df.iterrows(): 
        if (row['genero'] == gen):
            s_ = pd.DataFrame({
                col : row[col].split(sep)
            }) 
            summary = pd.concat( [ summary , s_ ] )
            summary['cant_'+gen] = 1
    summary = summary.groupby( [col] , as_index = False ).agg({ 'cant_'+gen : ["count"] })
    summary.columns = summary.columns.droplevel(1)
    return summary.sort_values(['cant_'+gen] , ascending=False )
```

```python
def get_freq_cleaning_characters(df, col, sep=";"):
    summary = pd.DataFrame()
    for index, row in df.iterrows():    
        s_ = pd.DataFrame({
            col : row[col].split(sep)
        }) 
        summary = pd.concat( [ summary , s_ ] )
        summary['n'] = 1
    summary = summary.groupby( [col] , as_index = False ).agg({ "n" : ["count"] })
    summary.columns = summary.columns.droplevel(1)
    return summary.sort_values(['n'] , ascending=False )
```

```python
get_freq_cleaning_characters(dfFuman,"acompanante",";")
```

```python
datosLimpios = dfFuman[(dfFuman['acompanante'] == "Solo") 
              | (dfFuman['acompanante'] == "Amigos") 
              | (dfFuman['acompanante'] == "Conocidos") 
              | (dfFuman['acompanante'] == "Familiares")
              | (dfFuman['acompanante'] == "Desconocidos")
             ]

otrosDatos = dfFuman[(dfFuman['acompanante'] != "Solo") 
              & (dfFuman['acompanante'] != "Amigos") 
              & (dfFuman['acompanante'] != "Conocidos") 
              & (dfFuman['acompanante'] != "Familiares")
              & (dfFuman['acompanante'] != "Desconocidos")
             ]

otrosDatos.loc[:,["acompanante"]]
contigency = pd.crosstab(datosLimpios["acompanante"], datosLimpios["fuma"])

contigency.loc["Varios"] = otrosDatos.shape[0]
contigency
```

```python
sns.heatmap( contigency , annot = True , cmap="YlGnBu" , fmt=".4g" )
```

```python
c, p, dof, expected = chi2_contingency(contigency)
p < 0.05
```

Si el p-value es menor al 0.05 las variables son independientes

si es p > 0.05: Por lo tanto no rechazamos la hipotesis nula (son independientes)

```python
print_bars(datosLimpios, "acompanante", "fuma")
print_bars(otrosDatos, "acompanante", "fuma")
```
